```python
# data engineer
links =[]
print('Links are being collected now.')

for page in range (1,40):
    get_url = driver.current_url
    response = requests.get(driver.current_url)
    soup = BeautifulSoup(response.content, 'html.parser')
    for link in soup.find_all('a'):
        if  str(link.get('href')).startswith("https://uk.linkedin.com/jobs/view/") and link.get('href') not in links:
                links.append(link.get('href'))
    print(f'Collecting links in the page: {page-1}')
    driver.find_element(By.XPATH,f"//button[@aria-label='Page {page}']").click()
    time.sleep(3)

print('Found ' + str(len(links)) + ' links for job offers')
```
